#  Исследование результатов А/В-теста и поиск инсайтов

SollmaFin — международное мобильное приложение для онлайн-торговли финансовыми активами. Пользователи могут инвестировать деньги в акции, валюту, криптовалюту, биржевые фонды (ETF) и другие активы. Целевая аудитория приложения — начинающие инвесторы. Получить доступ к финансовым рынкам можно со смартфона либо через веб-версию. 

Приложение ориентировано на рынок Латинской Америки с акцентом на четыре страны: Мексику, Бразилию, Колумбию и Аргентину — там быстро растёт интерес к финтеху и инвестициям, а экономики очень разные. 

Появилась гипотеза о недостаточной финансовой грамотности пользователей: они не понимают, как работать с активами разной степени риска. Клиенты покупают активы с высоким риском и теряют деньги, что снижает вовлечённость.
Команда продукта решила обновить онбординг пользователей и добавить в него детальную информацию о различиях финансовых активов и связанных с ними рисках. Однако появилось опасение, что углублённый онбординг отпугнёт клиентов от пополнения депозита и более рискованных вложений. 

Чтобы разобраться, команда решила провести A/B-эксперимент. 

### А/В-эксперимент

В эксперименте участвовали новые пользователи, которые зарегистрировались в приложении со 2 по 15 июня 2025 года. 
Пользователей случайным образом разделили на две равные группы: 
- Контрольная группа проходила стандартный онбординг без обязательного обучения.
- Тестовая группа проходила обновлённый онбординг с подробной информацией об активах и связанных с ними рисках. 

После разделения активность пользователей анализировали в течение недели.

Цель эксперимента — оценить, как обновлённый онбординг влияет на поведение пользователей и их дальнейшую инвестиционную активность.

При внедрении новой фичи команда продукта выдвинула такие гипотезы:
- Гипотеза роста: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
- Гипотеза риска: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.
- Дополнительная гипотеза: после нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит. При старом онбординге пользователи часто покупали активы с высоким риском без понимания последствий. Это приводило к потерям и оттоку после первого депозита. 

Чтобы увидеть полную картину, недостаточно следить за метриками конверсии. Команда выбрала для эксперимента такой набор метрик: 
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Ожидалось, что ключевая и барьерная метрики не упадут, а вспомогательные покажут значительный рост.

**Задача 2.2.** Анализ воронок событий.
- Постройте классическую и пошаговую воронки;
- Охарактеризуйте пользовательский путь в приложении и определите узкие места воронки. 


````python
steps = ['install / open_web', 'introduction', 'registration', 'main_page', 'onboarding_complete', 'first_deposit', 'asset_purchase', 'second_deposit']

df_event = df_sol_hist[df_sol_hist['event_name'].isin(steps)].copy()
df_counts = df_event.groupby('event_name')['user_id'].nunique().reindex(steps)

full_df = pd.DataFrame({
   'step': df_counts.index,
   'users': df_counts.values
})

full_df['conversion_first_%'] = round((full_df['users'] / full_df['users'][0]) * 100, 1)
full_df['conversion_previous_%'] = round((full_df['users'] / full_df['users'].shift(1)) * 100, 1)
````

````python
plt.figure(figsize=(12, 5))

plt.bar(full_df['step'], full_df['conversion_first_%'])

plt.title('Классическая воронка: конверсия от первого шага в процентах')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')

plt.xticks(rotation= 25)

plt.grid()
plt.show()
````

На основе предоставленного графика можно сделать следующие выводы о пользовательском пути в приложении и определить узкие места воронки.

Последовательность этапов следующая: `install / open_web`100% → `introduction`100% → `registration`93% → `main_page`85% → `onboarding_complete`84% → `first_deposit`67% → `asset_purchase`37% → `second_deposit`13%

- Главная страница (main_page): Падение с 90% до 80% может указывать на проблемы с интерфейсом или отсутствием интереса у пользователей.
- Первая покупка (first_deposit): Падение с 80% до 40% указывает на значительные проблемы с процессом покупки или на отсутствие мотивации у пользователей.
- Вторая покупка (second_deposit): Падение с 30% до 10% указывает на то, что большинство пользователей не возвращаются для повторных покупок, что может быть связано с неудовлетворенностью продуктом или сервисом.



````python
plt.figure(figsize=(12, 5))

plt.bar(full_df['step'][1:], full_df['conversion_previous_%'][1:])

plt.title('Step-by-step воронка: конверсия от предыдущего шага')
plt.ylabel('Конверсия от предыдущего шага (%)')
plt.xlabel('Этапы воронки')

plt.xticks(rotation= 25)

plt.grid()
plt.show()
````

Последовательность этапов следующая: `introduction`100% → `registration`93% → `main_page`92% → `onboarding_complete`98% → `first_deposit`80% → `asset_purchase`55% → `second_deposit`35%

Узкие места воронки можно определить как этапы, где происходит значительное снижение конверсии. В данном случае это этапы "asset_purchase" и "second_deposit". Эти этапы требуют дополнительного анализа и оптимизации для повышения конверсии и улучшения пользовательского опыта.







````python
platform_mob = df_sol_hist[df_sol_hist['platform'] == 'mobile'].copy()
platform_web = df_sol_hist[df_sol_hist['platform'] == 'web'].copy()
````

````python
df_counts_mob = platform_mob.groupby('event_name')['user_id'].nunique().reindex(steps)
df_counts_web = platform_web.groupby('event_name')['user_id'].nunique().reindex(steps)

full_df_mob = pd.DataFrame({
   'step': df_counts_mob.index,
   'users': df_counts_mob.values
})
full_df_web = pd.DataFrame({
   'step': df_counts_web.index,
   'users': df_counts_web.values
})

full_df_mob['conversion_previous_%'] = round((full_df_mob['users'] / full_df_mob['users'].shift(1)) * 100, 1)
full_df_web['conversion_previous_%'] = round((full_df_web['users'] / full_df_web['users'].shift(1)) * 100, 1)
````

````python
full_df_platform = full_df_web.merge(full_df_mob, on= 'step')
````

````python
full_df_platform[1:].plot(kind= 'bar', figsize= (12, 5), x= 'step', y= ['conversion_previous_%_x', 'conversion_previous_%_y'])

plt.title('Step-by-step воронка: конверсия от предыдущего шага (platform)')
plt.ylabel('Конверсия от предыдущего шага (%)')
plt.xlabel('Этапы воронки')

plt.xticks(rotation= 25)

plt.legend(['web', 'mobile'])
plt.grid()
plt.show()
````

Диаграмма предоставляет ценную информацию о конверсии воронки продаж для веб и мобильной платформ. Она помогает выявить сильные и слабые стороны каждого этапа и может быть использована для улучшения пользовательского опыта и увеличения конверсии.

````python
full_df_country = pd.DataFrame()

for i in df_sol_hist['country_code'].unique():
  df_cou = df_sol_hist[df_sol_hist['country_code'] == i]

  df_group = df_cou.groupby('event_name')['user_id'].nunique().reindex(steps)

  full_df_cou = pd.DataFrame({
    'step': df_group.index,
    'users': df_group.values
  })

  full_df_cou['conversion_previous_%'] = round((full_df_cou['users'] / full_df_cou['users'].shift(1)) * 100, 1)
  full_df_cou['country_code'] = i

  full_df_country = pd.concat([full_df_country, full_df_cou])
````

- В тестовой группе наблюдается более высокая вероятность открытия второго депозита при высоком и низком риске по сравнению с контрольной группой.
- В тестовой группе наблюдается более низкая вероятность открытия второго депозита при среднем риске по сравнению с контрольной группой.

Для более точного анализа и оценки статистической значимости необходимо провести дополнительные статистические тесты.

- Нулевая гипотеза (H₀): Конверсия в тестовой группе равна конверсии в контрольной группе (нет статистически значимой разницы между группами).
- Альтернативная гипотеза (H₁): Конверсия в тестовой группе отличается от конверсии в контрольной группе (есть статистически значимая разница).

Уровень значимости: α = 0.05. Если p-value < 0.05, отвергаем H₀ в пользу H₁.

````python
df_compare = conversion_cont.merge(conversion_test, on='risk_level', suffixes=('_cont', '_test'))
````

````python
from statsmodels.stats.proportion import proportions_ztest

def z_test_proportions(success_a, size_a, success_b, size_b):
    count = [success_a, success_b]
    nobs = [size_a, size_b]
    stat, pval = proportions_ztest(count, nobs)
    return stat, pval
````

````python
results = []

for _, row in df_compare.iterrows():
    # Количество успехов (вторых депозитов)
    success_cont = int(row['conversion_rate_cont'] * row['total_users_cont'])
    success_test = int(row['conversion_rate_test'] * row['total_users_test'])
    
    stat, pval = z_test_proportions(success_cont, row['total_users_cont'], success_test, row['total_users_test'])
    
    results.append({
        'risk_level': row['risk_level'],
        'conversion_cont': row['conversion_rate_cont'],
        'conversion_test': row['conversion_rate_test'],
        'z_stat': stat,
        'p_value': pval,
        'significant': pval < 0.05
    })

df_results = pd.DataFrame(results)
print(df_results)
````
